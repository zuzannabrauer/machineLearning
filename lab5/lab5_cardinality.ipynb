```python
import pandas as pd
import numpy as np
```

```python
titanic_data = pd.read_csv('https://raw.githubusercontent.com/zuzannabrauer/machineLearning/master/lab5/titanic_data.csv', skiprows = 17, index_col = False, names = ['pclass', 'survived', 'name', 'sex', 'age', 'sibsp', 'parch', 'ticket', 'fare', 'cabin', 'embarked', 'boat', 'body', 'home.dest'])
titanic_data = titanic_data.replace('?', np.NaN)
titanic_data
```

<h3>Oznaczenia kolumn w tabeli z danymi</h3>

* pcalss - numer klasy, w której podróżował pasażer
* survived - czy dana osoba przeżyła (0 = Nie, 1 = Tak)
* name - imię i nazwisko
* sex - płeć
* age - wiek w momencie zaokrętowania
* sibsp - ilość rodzenśtwa/partenrów pasażera obecnych na pokładzie
* parch - ilość rodziców/dzieci pasażera obecnych na pokładzie
* ticket - numer biletu
* fare - opłata za bilet (wyrażona w funtach brytyjskich)
* cabin - numer identyfikacyjny kabiny
* embarked - port, w którym pasażer się zaokrętował (C = Cherbourg; Q = Queenstown; S = Southampton)
* boat - numer identyfikacyjny łodzi ratunkowej
* body - numer identyfikacyjny ciała
* home.dest - destynacja pasażera

```python
for column in titanic_data:
  print('Liczba etykiet zmiennej '  + column + ': ' + format(len(titanic_data[column].unique())))
```

```python
print('Liczba wszystkich pasażerów: ' + format(len(titanic_data['name'].unique())))
```

<h3>Podział zmiennych ze względu na kardynalność</h3>
Poniższa klasyfikacja przedstawia podział zmiennych na grupy według mocy ich zbiorów (kardynalności).Jak możemy zauważyć, więcej zmiennych posiada dużą kardynalność. Przy podziale natknęłam się na wątpliwości dotyczące zmiennej boat, jej moc wynosi tylko 28 co nie jest ani małą ani dużą wartością (gdyby takich zmiennych było więcej to może potrzebna by była trzecia kategoria ze zbiorami o średniej mocy ?), jednak końcowo podjęłam decyzję o jej klasyfikacji do grupy o małej kardynalności.

1. Zmienne o małej mocy zbioru:
* pclass
* survived
* sex
* sibsp
* parch
* embarked

2. Zmienne o dużej mocy zbioru:
* name
* age
* ticket
* fare
